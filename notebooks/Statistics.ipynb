```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/cicy4folds_extended.csv')
df = df.replace('Null', np.nan)
nan_indeces = df.index[df.isna().any(axis=1)].tolist()

df_clear = df.drop(nan_indeces)
```

```python
def print_statistics(target_name):
    print(f"Number of unique hodge numbers: {df_clear[target_name].nunique()}")
    print(f"Minimum hodge number: {df_clear[target_name].min()}")
    print(f"Maximum hodge number: {df_clear[target_name].max()}")
```

```python
def hist_show(target_name):
    plt.title(f"Histogram of {target_name}")
    plt.hist(df_clear[target_name], df_clear[target_name].nunique())
    plt.show()
```

```python
target_name = 'h11'
print_statistics(target_name)
hist_show(target_name)
```
